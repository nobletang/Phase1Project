```python
# import relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in the csv file for title.basics
basics_df = pd.read_csv('zippedData/imdb.title.basics.csv.gz')
```

```python
basics_df.info()
```

```python
# read in the csv file for the title.ratings
gross_df = pd.read_csv('zippedData/bom.movie_gross.csv')
gross_df.info()
```

```python
gross_df.dropna(inplace=True)
gross_df.info()
```

```python
gross_df['foreign_gross'] = gross_df['foreign_gross'].str.replace(',', '')
gross_df['foreign_gross'] = gross_df['foreign_gross'].astype(float)
gross_df.info()
```

```python
# read in the csv file for the rt_movie_info
budgets_df = pd.read_csv('zippedData/tn.movie_budgets.csv')
```

```python
budgets_df.info()
```

```python
budgets_df.head()
```

```python
budgets_df.info()
```

```python
# looks like these columns are objects. Let's convert to integers and find some correlations

# production_budget
budgets_df['production_budget'] = budgets_df['production_budget'].str.replace(',', '')
budgets_df['production_budget'] = budgets_df['production_budget'].str.replace('$', '')
budgets_df['production_budget'] = budgets_df['production_budget'].astype(int)
budgets_df['production_budget']
```

```python
# domestic_gross
budgets_df['domestic_gross'] = budgets_df['domestic_gross'].str.replace(',', '')
budgets_df['domestic_gross'] = budgets_df['domestic_gross'].str.replace('$', '')
budgets_df['domestic_gross'] = budgets_df['domestic_gross'].astype(int)
budgets_df['domestic_gross']
```

```python
# worldwide_gross
budgets_df['worldwide_gross'] = budgets_df['worldwide_gross'].str.replace(',', '')
budgets_df['worldwide_gross'] = budgets_df['worldwide_gross'].str.replace('$', '')
budgets_df['worldwide_gross'] = budgets_df['worldwide_gross'].astype(int)
budgets_df['worldwide_gross']
```

## Make sure that they are stored as integers now

```python
budgets_df.info()
```

```python
# convert the release date to datetime object
# possibility that there is a correlation between the success and the time of release
budgets_df['release_date'] = pd.to_datetime(budgets_df['release_date'])
budgets_df['release_date']
```

```python
# now let's check our data types
budgets_df.info()
```

```python
# correlation
budgets_df.corr()
```

# I will try to extract seasons out of the datetime objects to see if there are ideal times of the year to release a movie based on their domestic and worldwide gross

```python
budgets_df['release_date'][0].month
```

```python
budgets_df['season'] = (budgets_df['release_date'].dt.month%12 + 3)//3

seasons = {
             1: 'Winter',
             2: 'Spring',
             3: 'Summer',
             4: 'Autumn'
}

budgets_df['season_name'] = budgets_df['season'].map(seasons)
```

```python
budgets_df.info()
```

```python
budgets_df.corr()
```

```python
budgets_df['season_name'].value_counts()
```

```python
autumn_ww = budgets_df[budgets_df['season_name'] == 'Autumn']['worldwide_gross'].sum()
winter_ww = budgets_df[budgets_df['season_name'] == 'Winter']['worldwide_gross'].sum()
spring_ww = budgets_df[budgets_df['season_name'] == 'Spring']['worldwide_gross'].sum()
summer_ww = budgets_df[budgets_df['season_name'] == 'Summer']['worldwide_gross'].sum()

print(f'The total worldwide_gross in Autumn is: {autumn_ww}')
print(f'The total worldwide_gross in Winter is: {winter_ww}')
print(f'The total worldwide_gross in Spring is: {spring_ww}')
print(f'The total worldwide_gross in Summer is: {summer_ww}')
```
