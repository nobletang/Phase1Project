```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tmbd_df = pd.read_csv('zippedData/tmdb.movies.csv', index_col=0)
tmbd_df.head()
```

```python
gross_df = pd.read_csv('zippedData/bom.movie_gross.csv')
gross_df.head()
```

```python
budgets_df = pd.read_csv('zippedData/tn.movie_budgets.csv')
budgets_df.head()
```

```python
# convert 'movie' to title so I can add them into a single
# dataframe together
budgets_df['title'] = budgets_df['movie']
budgets_df.drop('movie', axis=1, inplace=True)
```

```python
budgets_df.info()
```

```python
tmbd_budgets_df = tmbd_df.merge(gross_df, how='left', on='title')
tmbd_budgets_df.info()
```

```python
master_df = tmbd_budgets_df.merge(budgets_df, how='left', on='title')
master_df.info()
```

```python
# drop missing values
master_df.dropna(inplace=True)
```

```python
master_df.info()
```

```python
master_df.head()
```

# Drop OR convert the gross columns to integers (we have duplicate columns)

```python
# drop domestic gross x
master_df = master_df.drop('domestic_gross_x', axis=1)
master_df
```

```python
# drop one of the release date columns
master_df = master_df.drop('release_date_y', axis=1)
master_df
```

```python
# dropping other columns we don't need
master_df = master_df.drop(['genre_ids','id_x','studio','year','id_y'], axis=1)
```

```python
master_df.head()
```

```python
master_df.drop('original_title', axis=1, inplace=True)
master_df.info()
```

# Jobs: 1) Convert the gross columns to integers 2) Convert the release_date_x to datetime

```python
# convert release date x to datetime
master_df['release_date'] = pd.to_datetime(master_df['release_date_x'])
master_df
```

```python
# drop release_date_x
master_df.drop('release_date_x', axis=1, inplace=True)
master_df.info()
```

## Time to convert the gross columns to integers or floats

```python
master_df['production_budget'] = master_df['production_budget'].str.replace(',', '')
master_df['production_budget'] = master_df['production_budget'].str.replace('$', '')
master_df['production_budget'] = master_df['production_budget'].astype(int)
master_df['production_budget']
```

```python
master_df['domestic_gross'] = master_df['domestic_gross_y'].str.replace(',','')
master_df['domestic_gross'] = master_df['domestic_gross_y'].str.replace('$','')
master_df['domestic_gross'] = master_df['domestic_gross'].str.replace(',','')
master_df['domestic_gross'] = master_df['domestic_gross'].astype(int)
```

```python
master_df['domestic_gross']
```

```python
master_df['foreign_gross'] = master_df['foreign_gross'].str.replace(',', '')
master_df['foreign_gross'] = master_df['foreign_gross'].str.replace('$', '')
#master_df['foreign_gross'] = master_df['foreign_gross'].astype(int)
master_df['foreign_gross'] = master_df['foreign_gross'].astype(float)
```

```python
master_df['foreign_gross']
```

```python
master_df['worldwide_gross'] = master_df['worldwide_gross'].str.replace(',', '')
master_df['worldwide_gross'] = master_df['worldwide_gross'].str.replace('$', '')
master_df['worldwide_gross'] = master_df['worldwide_gross'].astype(int)
master_df['worldwide_gross']
```

# Correlation test

```python
sns.heatmap(master_df.corr(), annot=True)
```

```python
master_df.corr()
```

# Divide release dates into seasons

```python
master_df['season'] = (master_df['release_date'].dt.month%12 + 3)//3

seasons = {
             1: 'Winter',
             2: 'Spring',
             3: 'Summer',
             4: 'Autumn'
}

master_df['season_name'] = master_df['season'].map(seasons)
```

```python
master_df.corr()
```
